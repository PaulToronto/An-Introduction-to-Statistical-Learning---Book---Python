## Wage Data

```python
wages['education_level'] = wages['education'].str[0].astype('category')
wages['education_level']
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(8, 4))

sns.regplot(ax=ax[0],
            data=wages, x='age', y='wage', order=2,
            scatter_kws={'color':'dodgerblue', 'alpha':0.4, 's':20},
            line_kws={'color':'red'})
ax[0].set(xlabel = 'Age', ylabel='Wage')

sns.regplot(ax=ax[1],
            data=wages, x='year', y='wage',
            scatter_kws={'color':'dodgerblue', 'alpha':0.4, 's':20},
            line_kws={'color':'red'})
ax[1].set(xlabel = 'Year', ylabel='Wage')

sns.boxplot(ax=ax[2],
            data=wages, x='education_level', y='wage',
            flierprops={'marker':'o', 'markersize':3, 'markerfacecolor':'none'})
ax[2].set(xlabel='Education Level', ylabel='Wage')

plt.tight_layout()
plt.show()
```

```python
wages['education'].value_counts(dropna=False).sort_index()
```

## Stock Market Data

```python
smarket['Direction'] = smarket['Direction'].astype('category')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(8, 4))

sns.boxplot(ax=ax[0], data=smarket, x='Direction', y='Lag1',
            flierprops={'marker':'o', 'markersize':3, 'markerfacecolor':'none'})
ax[0].set(xlabel='Today\'s Direction', ylabel='Percentage change in S&P',
          title='Yesterday')

sns.boxplot(ax=ax[1], data=smarket, x='Direction', y='Lag2',
            flierprops={'marker':'o', 'markersize':3, 'markerfacecolor':'none'})
ax[1].set(xlabel='Today\'s Direction', ylabel='Percentage change in S&P',
          title='Two Days Previous')

sns.boxplot(ax=ax[2], data=smarket, x='Direction', y='Lag3',
            flierprops={'marker':'o', 'markersize':3, 'markerfacecolor':'none'})
ax[2].set(xlabel='Today\'s Direction', ylabel='Percentage change in S&P',
          title='Three Days Previous')

plt.tight_layout()
plt.show()
```
